```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import TimeoutException
import time, json, os, re
from bs4 import BeautifulSoup
```

```python
# 設定 Chrome 選項
options = webdriver.ChromeOptions()
# options.add_argument("--headless")  # 在背景執行，不開啟瀏覽器視窗
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36")

driver = None
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
driver.set_page_load_timeout(30)
```

```python
urls = []
```

```python
for i in range(1,11):
    fonfood_url = f"https://www.fonfood.com/store/153954/post/{i}"

    print(f"正在前往目標頁面: {fonfood_url}")
    driver.get(fonfood_url)

    # 使用 WebDriverWait 等待目標連結元素出現，最多等待 10 秒
    # 我們要找的是在 class="post-content" 區塊內的 <a> 標籤
    print("正在等待並尋找目標 URL...")
    wait = WebDriverWait(driver, 10)

    css_selector = f"a.postTitle"
    link_elements = driver.find_elements(By.CSS_SELECTOR, css_selector)
    print(f"找到 URL...{len(link_elements)}")

    for link_element in link_elements:
        # 取得 'href' 屬性值
        href = link_element.get_attribute('href')
        # 確保 href 存在且不為空
        if href:
            urls.append(href)
    time.sleep(5)
```

```python
len(urls)
```

```python
html_text = []
driver.set_page_load_timeout(60)

for u in urls:
    try:
        driver.get(u)
        html = driver.page_source
        soup = BeautifulSoup(html, "html.parser")
        for tag in soup(["script", "style", "header", "footer", "nav", "aside"]):
            tag.decompose()
        text = soup.get_text("\n")
        text = re.sub(r"\s+", " ", text)
        print(text[:300])
        html_text.append(text)
        time.sleep(1)
    except Exception as e:
        html = driver.page_source
        soup = BeautifulSoup(html, "html.parser")
        for tag in soup(["script", "style", "header", "footer", "nav", "aside"]):
            tag.decompose()
        text = soup.get_text("\n")
        text = re.sub(r"\s+", " ", text)
        print(text[:300])
        html_text.append(text)
        time.sleep(1)

```

```python
def extract_axia_related(text, client, model="gpt-4o-mini", max_chars=8000):
    """保留阿霞飯店相關敘述並盡量保留原描述語句。"""
    # 防止 token 過長，先截斷
    text = text[:max_chars]

    prompt = f"""
你是美食文案整理助手。請從下列內容中，
只保留與「阿霞飯店」直接相關的敘述（如菜色、作法、評價摘要等，不需要輸出地址電話），
移除與其他店家或不相關的段落。儘量保持原句與原用詞，不要虛構內容。
直接輸出整理後的文字，不要加任何前言或結語。

【原始內容】
{text}
"""

    resp = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": "你會做可信的文字過濾與摘錄，不添加無法驗證的資訊。"},
            {"role": "user", "content": prompt},
        ],
        temperature=0.2,
    )
    return (resp.choices[0].message.content or "").strip()

# 用法
resp_list = []
for h in html_text:
    try:
        out = extract_axia_related(h, client)
        print(out[:300])
        resp_list.append(out)
    except Exception as e:
        print("LLM 錯誤：", e)
        resp_list.append(f"[ERROR] {e}")
```
